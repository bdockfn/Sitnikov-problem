# Sitnikov problem



This program solve [Sitnikov problem](https://en.wikipedia.org/wiki/Sitnikov_problem)


initial data: $z_0, v_0, a, e$ \
$t \rightsquigarrow E$ \
$r = a*(1-e*cos(E))$ 

$\frac{dz}{dE} = 2rv$ 

$\frac{dv}{dE} = \frac {-2rz} {(\sqrt{z^2 + r^2})^3}$

$z \to  \frac{dz}{dE}$;
$v \to  \frac{dv}{dE}$

```python
import numpy as np
import matplotlib.pyplot as plt
from sitnikov_module import *
%matplotlib widget
```

##### Initial data

*a*: semi-major axis $a$ \
*e*: eccentricity $e$ \
*N*: number of periods \
*period*: value of one period \
*time_step*: step for values of ecccentric anomaly $E$


Making boundaries for defining the initial data even grid: \
*start_z, stop_z, start_v, stop_v, grid_step*


```python
a = 1/2
e = 0.1
N = 100
period = 2*np.pi
time_step = 0.01

start_z = -5
stop_z = 5
start_v = -2
stop_v = 2
grid_step = 0.5
```

Calculations:

```python
y0 = initial_grid_zv(start_z, stop_z, start_v, stop_v, grid_step)
#for k in range(len(y0[:,1])):    y0[k,1]=0
res = solution(y0, a, e, period, N, time_step)
```

Plotting:


```python
def show(res):
    z = res[:,0]
    v = res[:,1]

    fig = plt.figure(figsize = (6, 5))
    plt.scatter(z, v, color = 'red', s = 0.03)
    plt.xlim(-12, 12)
    #plt.ylim(-2, 2)
    plt.grid()
    plt.show()
show(res)
```
